```python
import yaml

data = {
    "path": "/kaggle/input/datasets/mrunalkalebtech2023/ppe-augmented-dataset/dataset",

    "train": "train/images",
    "val": "val/images",
    "test": "test/images",

    "names": {
        0: "helmet",
        1: "vest",
        2: "gloves",
        3: "glasses"
    }
}

with open("data.yaml", "w") as f:
    yaml.dump(data, f)

print("data.yaml created")
```

```python
from ultralytics import YOLO

#trained model
model = YOLO("/kaggle/input/models/mrunalkalebtech2023/model1/tensorflow2/default/1/best.pt")  # path to your weights


metrics = model.val(data="data.yaml", split="val")


print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
```

```python
from ultralytics import YOLO

metrics = model.val(data="data.yaml")

names = model.names

# Per-class AP (mAP50-95)
ap = metrics.box.ap  

# Per-class AP@0.5
ap50 = metrics.box.ap50

for i, name in names.items():
    print(f"Class: {name}")
    print(f"  AP50: {ap50[i]:.4f}")
    print(f"  AP50-95: {ap[i]:.4f}")
```
